# Displaying a 2d array as an image

## Monochrome


```python
m = np.random.randint(0, 2, (5,5), dtype=bool)
m
```

```python
plt.imshow(m);  # popular function for visualization of 2-dimenstion arrays
# The ; is for displaying cleaner output in Jupyter - dont display the result of the last expression
```

## Scalar data, Normalized

```python
a = np.arange(100).reshape(10, 10)
```

```python
plt.imshow(a);
```

```python
plt.imshow(a)
plt.colorbar();
```

```python
plt.imshow(a, 'gray')
plt.colorbar();
```

```python
plt.imshow(a, 'Reds')
plt.colorbar();
```

```python
plt.imshow(a, plt.cm.Reds)
plt.colorbar();
```

## Indexing Example: RGB data

```python
m = np.random.random((150, 200, 3))  # if we give 3-dim arrays instead of 2, it assumes the data is RGB data
```

```python
m
```

```python
plt.imshow(m);
```

```python
m[61, 110] # one pixel
```

```python
m[25:30, 100:104] # a 5x4x3 view
```

```python
m[:,:,0] # The R channel, 150x200 view
```

```python
plt.imshow(m[:,:,0], vmin=0, vmax=1)
plt.colorbar();
```

```python
plt.imshow(m[:,:,0], 'gray', vmin=0, vmax=1)
plt.colorbar();
```
