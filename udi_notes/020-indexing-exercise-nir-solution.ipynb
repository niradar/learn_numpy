## Part 1: Show the image flipped vertically

```python
plt.imshow(a[::-1,:])
```

## Part 2: Show the image flipped horizontally

```python
plt.imshow(a[:, ::-1])
```

## Part 3: Show channels r,g and b 


```python
for i, c in enumerate(('Reds', 'Greens', 'Blues')):
    plt.imshow(a[:,:,i], c, vmin=0, vmax=255)
    # a[:,:,0] means take the first element of the rgb, a[:,:,1] is the second, etc
    plt.colorbar()
    plt.title(c)
    plt.show()
```
